```python
import numpy as np
from PIL import Image

image_path = "D:/IIT/SOC/Week 2/Assignment1/a.png"

img = Image.open(image_path)

img_array = np.array(img)
gray_values = img_array.ravel()

min_value = np.min(gray_values)
max_value = np.max(gray_values)

stretched_values = (gray_values - min_value) * (255 / (max_value - min_value))

img_array = stretched_values.reshape(img_array.shape)

img_stretched = Image.fromarray(np.uint8(img_array))

final_image_path = "D:/IIT/SOC/Week 2/Assignment1/a_output.png"
img_stretched.save(final_image_path)
```
